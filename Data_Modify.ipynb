```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv("Iphone_clean.csv")
df.head()
```

```python
df.drop('reviewedIn', axis=1, inplace= True)
df.head()
```

```python
different_products = df['productAsin'].unique()
different_products
```

```python
different_variants = df['variantAsin'].unique()
different_variants
```

```python
# Function to extract Color and Size using regex
def extract_color_size(attribute):
    match = re.search(r'Colour:\s*(\w+)(?=Size:)\s*Size:\s*(\d+\s*GB)', attribute)
    if match:
        return match.groups()  # Returns a tuple (color, size)
    return None, None  # In case of no match

# Splitting the 'variant' column into 'Color' and 'Size'
df[['Color', 'Size']] = df['variant'].apply(lambda x: pd.Series(extract_color_size(x.replace("Size:", "Size: "))))

```

```python
df.drop('variant', axis=1, inplace=True)
df.head()
```

```python
country = df["country"].unique()
country
```

```python
df.drop("reviewDescription", axis=1, inplace=True)
```
